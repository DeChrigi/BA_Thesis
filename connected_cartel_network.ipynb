```python
import pandas as pd
import numpy as np
```

```python
# read cartel excel file
cartel_df = pd.read_excel('./transformed_data/cartel_networks/cartel_data_with_centrality_measures.xlsx', sheet_name='Sheet1')

cartel_df.head(10)
```

```python
# select only relevant columns
cartel_df = cartel_df[['case', 'cartel', 'entity_name', 'sector', 'subsector', 'start', 'end', 'duration']]

# Select unique rows
cartel_df = cartel_df.drop_duplicates()

# define new column in dataframe year_minus_3
cartel_df['t_A_-3'] = cartel_df['start'] - 3
cartel_df['t_A_-2'] = cartel_df['start'] - 2
cartel_df['t_A_-1'] = cartel_df['start'] - 1

cartel_df['t_A'] = cartel_df['start']
cartel_df['t_AE'] = cartel_df['start'].astype(str) + ' - ' + cartel_df['end'].astype(str)
cartel_df['t_E'] = cartel_df['end']

cartel_df['t_E_+1'] = cartel_df['end'] + 1
cartel_df['t_E_+2'] = cartel_df['end'] + 2
cartel_df['t_E_+3'] = cartel_df['end'] + 3

# get smallest value from column t_A_-3
min_t_a_minus_3 = cartel_df['t_A_-3'].min()

# get largest value from column t_E_+3
max_t_e_plus_3 = cartel_df['t_E_+3'].max()

print( min_t_a_minus_3)
print(max_t_e_plus_3)

cartel_df.head(10)
    
```

```python
# Get all filenames from mapping file (contains all filenames)
df_mapping = pd.read_excel("./raw_data/mapping/filename_to_entity_mapping.xlsx")

# if entity_name_direct_match column is not empty take entity_name_direct_match else take entity_name_indirect_match_1
df_mapping['entity_name'] = df_mapping.apply(
    lambda x: x['entity_name_direct_match'] if pd.notna(x['entity_name_direct_match']) else x['entity_name_indirect_match_1'], axis=1)

# drop all columns except entity_name and filename
df_mapping = df_mapping[['entity_name', 'file_name']]

# join cartel_df with df_mapping on entity_name
cartel_df = cartel_df.merge(df_mapping, on='entity_name', how='left')

# drop all rows where file_name is empty
cartel_df = cartel_df[cartel_df['file_name'].notna()]

# count number of rows in cartel_df
num_rows = cartel_df.shape[0]

print(f"Number of rows in cartel_df after mapping: {num_rows}")

cartel_df.head(10)
```

```python
herfindal_df = pd.read_excel('./transformed_data/shareholder_data/herfindahl_index.xlsx', sheet_name='Sheet1')

def get_herfindahl_value(year, company_name):
    filtered = herfindal_df.loc[(herfindal_df['year'] == year) & (herfindal_df["company_name"] == company_name), 'herfindahl_index']
    return filtered.values[0] if not filtered.empty else np.nan

# for each row in cartel_df
for index, row in cartel_df.iterrows():

    file_name = row['file_name']
    t_a_minus_3 = row['t_A_-3']
    t_a_minus_2 = row['t_A_-2']
    t_a_minus_1 = row['t_A_-1']
    t_a = row['t_A']
    t_a_e = row['t_AE']
    t_e = row['t_E']
    t_e_plus_1 = row['t_E_+1']
    t_e_plus_2 = row['t_E_+2']
    t_e_plus_3 = row['t_E_+3']




    t_a_minus_3_v = get_herfindahl_value(t_a_minus_3, file_name)
    t_a_minus_2_v = get_herfindahl_value(t_a_minus_2, file_name)
    t_a_minus_1_v = get_herfindahl_value(t_a_minus_1, file_name)
    t_a_v = get_herfindahl_value(t_a, file_name)
    
    # t_a_e_v is the average of the years between t_a and t_e
    t_a_e_v = herfindal_df.loc[(herfindal_df['year'] >= t_a) & (herfindal_df['year'] <= t_e) & (herfindal_df["company_name"] == file_name), 'herfindahl_index'].mean()

    t_e_v = get_herfindahl_value(t_e, file_name)
    t_e_plus_1_v = get_herfindahl_value(t_e_plus_1, file_name)
    t_e_plus_2_v = get_herfindahl_value(t_e_plus_2, file_name)
    t_e_plus_3_v = get_herfindahl_value(t_e_plus_3, file_name)

    # set the values in cartel_df
    cartel_df.at[index, 't_A_-3_v'] = t_a_minus_3_v
    cartel_df.at[index, 't_A_-2_v'] = t_a_minus_2_v
    cartel_df.at[index, 't_A_-1_v'] = t_a_minus_1_v
    cartel_df.at[index, 't_A_v'] = t_a_v
    cartel_df.at[index, 't_AE_v'] = t_a_e_v
    cartel_df.at[index, 't_E_v'] = t_e_v
    cartel_df.at[index, 't_E_+1_v'] = t_e_plus_1_v
    cartel_df.at[index, 't_E_+2_v'] = t_e_plus_2_v
    cartel_df.at[index, 't_E_+3_v'] = t_e_plus_3_v

cartel_df.head(10)

```

```python
# save cartel_df to excel file
cartel_df.to_excel('./transformed_data/cartel_networks/cartel_data_with_herfindahl_index.xlsx', index=False)
```

```python
# Get all filenames from mapping file (contains all filenames)
df_mapping = pd.read_excel("./raw_data/mapping/filename_to_entity_mapping.xlsx")

# if entity_name_direct_match column is not empty take entity_name_direct_match else take entity_name_indirect_match_1
df_mapping['entity_name'] = df_mapping.apply(
    lambda x: x['entity_name_direct_match'] if pd.notna(x['entity_name_direct_match']) else x['entity_name_indirect_match_1'], axis=1)

# drop all columns except entity_name and filename
df_mapping = df_mapping[['entity_name', 'file_name']]

# join cartel_df with df_mapping on entity_name
cartel_df_v2 = cartel_df_v2.merge(df_mapping, on='entity_name', how='left')

# count number of rows in cartel_df
print(f"number of records in cartel_df_v2:  {cartel_df_v2.shape[0]}")

cartel_df_v2.head(10)
```

```python
# Assuming your DataFrame is named df
# Step 1: Filter non-null file names
cartel_df_v2_non_null = cartel_df_v2[cartel_df_v2['file_name'].notnull()]

# Step 2: Group by 'case' and 'cartel', and count non-null file_name entries
cartel_df_v2_fn_counts = cartel_df_v2_non_null.groupby(['case', 'cartel'])['file_name'].count().reset_index()

# Optional: Rename the column to something meaningful
cartel_df_v2_fn_counts = cartel_df_v2_fn_counts.rename(columns={'file_name': 'number_of_firms_with_file'})

cartel_df_v2_fn_counts.head(10)
```

```python
cartel_df_v2 = cartel_df_v2.merge(cartel_df_v2_fn_counts, left_on=['case', 'cartel'], right_on=['case', 'cartel'], how='left')

# fill NaN values with 0
cartel_df_v2['number_of_firms_with_file'] = cartel_df_v2['number_of_firms_with_file'].fillna(0)
# convert to int
cartel_df_v2['number_of_firms_with_file'] = cartel_df_v2['number_of_firms_with_file'].astype(int)

# fill NaN values with 0
cartel_df_v2['number_of_cartel_firms'] = cartel_df_v2['number_of_cartel_firms'].fillna(0)
# convert to int
cartel_df_v2['number_of_cartel_firms'] = cartel_df_v2['number_of_cartel_firms'].astype(int)

# fill NaN values with 0
cartel_df_v2['number_of_firms'] = cartel_df_v2['number_of_firms'].fillna(0)
# convert to int
cartel_df_v2['number_of_firms'] = cartel_df_v2['number_of_firms'].astype(int)

# change order of columns
cartel_df_v2 = cartel_df_v2[['case', 'cartel', 'number_of_firms', 'number_of_cartel_firms', 'number_of_firms_with_file', 'sector', 'subsector', 'entity_name', 'file_name', 'duration', 'start', 'end', ]]

print(f"number of records in cartel_df_v2:  {cartel_df_v2.shape[0]}")

print(cartel_df_v2.dtypes)

cartel_df_v2.head(10)
```
