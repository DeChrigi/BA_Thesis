```python
import pandas as pd
```

```python
# read cartel excel file
cartel_df = pd.read_excel('./transformed_data/cartel_networks/cartel_data_with_centrality_measures.xlsx', sheet_name='Sheet1')

cartel_df.head(10)
```

```python
# select only relevant columns
cartel_df = cartel_df[['case', 'cartel', 'sector', 'subsector', 'start', 'end', 'duration']]

# Select unique rows
cartel_df = cartel_df.drop_duplicates()

# define new column in dataframe year_minus_3
cartel_df['t_A_-3'] = cartel_df['start'] - 3
cartel_df['t_A_-2'] = cartel_df['start'] - 2
cartel_df['t_A_-1'] = cartel_df['start'] - 1

cartel_df['t_A'] = cartel_df['start']
cartel_df['t_E'] = cartel_df['end']

cartel_df['t_E_+1'] = cartel_df['end'] + 1
cartel_df['t_E_+2'] = cartel_df['end'] + 2
cartel_df['t_E_+3'] = cartel_df['end'] + 3

# get smallest value from column t_A_-3
min_t_a_minus_3 = cartel_df['t_A_-3'].min()

# get largest value from column t_E_+3
max_t_e_plus_3 = cartel_df['t_E_+3'].max()

print( min_t_a_minus_3)
print(max_t_e_plus_3)

cartel_df.head()
    
```
